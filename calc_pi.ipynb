```python
import numpy as np
import matplotlib.pyplot as plt
from control import *

#================================================= MOTOR LEFT
#=================================================
# CARACTERISTIQUES DU 1er ORDRE IDENTIFIE

G = 1332 / 50  # Gain du systme
tau = 230*10**-3  # Constante de temps
#=================================================
# SYSTEME A ASSERVIR ( EN BOUCLE OUVERTE )
FTBO = tf([G], [tau, 1]) 

print("Fonction de transfert en boucle ouverte :")
print(FTBO)

# Rponse  un chelon du systme
t, y = step_response(FTBO)
plt.figure()
plt.plot(t, y)
plt.grid()
plt.title("Rponse  un chelon du systme en boucle ouverte")
plt.xlabel('Temps (s)')
plt.ylabel('FTBO(t)')
plt.show()

#=================================================
# CORRECTEUR PI
ti = 0.1 * tau  # Temps d'intgration
Kpi = 0.01  # Gain proportionnel

CORR = Kpi * tf([ti, 1], [ti, 0])
print("Fonction de transfert du correcteur PI :")
print(CORR)

# Bode plot pour le correcteur PI
plt.figure()
bode(CORR, Hz=True, dB=True)
plt.title("Diagramme de Bode du correcteur PI")
plt.grid()
plt.show()

#=================================================
# ASSOCIATION CORRECTEUR PI + SYSTEME A ASSERVIR
FTBO_CORR = CORR * FTBO
gm, pm, wg, wp = margin(FTBO_CORR)

print(f"Marge de gain : {gm:.2f} dB")
print(f"Marge de phase : {pm:.2f}")
print(f"Frquence de gain : {wg:.2f} rad/s")
print(f"Frquence de phase : {wp:.2f} rad/s")

# Bode plot pour le systme avec correcteur
plt.figure()
bode(FTBO_CORR, dB=True)
plt.title("Diagramme de Bode du systme avec correcteur PI")
plt.grid()
plt.show()

#=================================================
# SYSTEME CORRIGE EN BOUCLE FERMEE
FTBF = (CORR * FTBO) / (1 + CORR * FTBO)

# Rponse  un chelon du systme corrig
t, y = step_response(FTBF)

plt.figure()
plt.plot(t, y)
plt.grid()
plt.title("Rponse  un chelon du systme en boucle ferme")
plt.xlabel('Temps (s)')
plt.ylabel('FTBF(t)')
plt.show()
#=================================================

#=================================================
# CORRECTEUR NUMERIQUE
# Dfinition des variables
Kp = 0.01  # Gain proportionnel
Ki = ti / tau  # Gain intgral
Te = 0.01  # Priode d'chantillonnage

# Initialisation des signaux
n = 100  # Nombre d'chantillons
up = np.zeros(n)  # Signal de contrle proportionnel
ui = np.zeros(n)  # Signal de contrle intgral
cde = np.zeros(n)  # Commande totale
epsilon = np.random.normal(0, 1, n)  # Erreur alatoire (peut tre remplace par la vritable erreur)

# Boucle de calcul
for i in range(1, n):
    up[i] = Kp * epsilon[i]  # Proportionnel
    ui[i] = ui[i - 1] + Kp * Ki * epsilon[i]  # Intgral
    cde[i] = up[i] + ui[i]  # Commande totale

# Affichage des rsultats
plt.figure()
plt.plot(cde, label='Commande Totale (cde[n])')
plt.plot(up, label='Commande Proportionnelle (up[n])')
plt.plot(ui, label='Commande Intgrale (ui[n])')
plt.grid()
plt.title("Commandes du correcteur numrique")
plt.xlabel('Echantillons')
plt.ylabel('Valeur de la commande')
plt.legend()
plt.show()
#=================================================
```
